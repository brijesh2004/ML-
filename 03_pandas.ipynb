Inspecting a DataFrame

```python
diabetes_df.tail()  # last five row of the data set
```

```python
diabetes_df.info()
```

```python
# finding the number of missing values

diabetes_df.isnull().sum()
```

```python
# counting the values based on the labels
diabetes_df.value_counts('Outcome')
```

```python
# group the values based on the mean
diabetes_df.groupby('Outcome').mean()
```

Statistical Measures

```python
# count or number of values

diabetes_df.count()  # number of values in each columns
```

```python
# mean values columns wise
diabetes_df.mean()
```

```python
# standard derivation - column wise
diabetes_df.std()
```

```python
# minimum value
diabetes_df.min()
```

```python
# maximum value

diabetes_df.max()
```

```python
# all the statistical measures about the dataframe
diabetes_df.describe()
```

Manipulating a DataFrame

```python
# adding a column to a dataframe

diabetes_df['Hello'] = np.random.rand(len(diabetes_df))
```

```python
diabetes_df.head()
```

```python
# removing a row

diabetes_df.drop(index=0 , axis=0)
```

```python
# drop a column
diabetes_df.drop(columns='Hello' , axis=1)
```

```python
# locating a row using the index value
diabetes_df.iloc[2]
```

```python
# locating a particular column
diabetes_df.iloc[:,0] # first column
```
