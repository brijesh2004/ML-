# Imbalanced Dataset:
1. A dataset with an unequal distribution

```python
import numpy as np
import pandas as pd
```

```python
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.shape
```

```python
# distribution of the two classes
credit_card_data['Class'].value_counts()
```

0--> Legit transactions

1 --> Fraudulent Transactions

This is highly imbalanced Dataset

```python
# seperating the legit and fraudulent transactions

legit = credit_card_data[credit_card_data.Class == 0.0]
fraud = credit_card_data[credit_card_data.Class == 1.0]
```

```python
legit
```

```python
print(legit.shape , fraud.shape)
```

Under-sampling

Building a sample dataset containing similar distribution of Legit & fraudulent transactions

Number of Fraudulent Transactions --> 3

```python
legit_sample = legit.sample(n=6)
```

```python
legit_sample
```

Concatenate the two DataFrames

```python
new_dataset = pd.concat([legit_sample , fraud] , axis = 0)
```

```python
new_dataset
```
