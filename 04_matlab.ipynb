# Matplotlib :
  . useful for making plots

```python
# importing matplotlib library

import matplotlib.pyplot as plt
```

```python
import numpy as np  # to get the data for our plots
```

```python
x = np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

plotting the data

```python
# sine wave

plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
```

```python
# adding title x-axis and -y axis labels

plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10 ,10,20)
y = x**2
plt.plot(x,y)
```

```python
plt.plot(x,y , 'r*')
```

```python
plt.plot(x,y , "b+")
```

```python
# multiple graph

x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x) , 'g--')
plt.plot(x , np.cos(x) , 'r-')
```

Bar plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English' , 'Hindi' , "French" , 'Spanish' , 'Latin']
people = [100 , 50,150,40,70]
ax.bar(languages , people)
plt.xlabel('Languages')
plt.ylabel('Number of People')
```

Pie chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English' , 'Hindi' , "French" , 'Spanish' , 'Latin']
people = [100 , 50,150,40,70]
ax.pie(people , labels=languages , autopct='%1.1f%%')  #1.1 means it is 1 point decimal value
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English' , 'Hindi' , "French" , 'Spanish' , 'Latin']
people = [100 , 50,150,40,70]
ax.pie(people , labels=languages , autopct='%1.3f%%')  #1.1 means it is 1 point decimal value
plt.show()
```

Scatter plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='green')
ax.scatter(x,z,color='black')
```

3D scatter plot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
```
